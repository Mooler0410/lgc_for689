```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_theme(style="darkgrid", palette = "dark")

data_pd = pd.read_csv('miu-sigma-64-500-remove.csv')

best_miu = data_pd[data_pd['Alpha']==0.3]
```

```python
sigma = [ f"{i}" for i in sorted(best_miu['Sigma'], reverse=True)]


best_miu = best_miu.sort_values(by=['Sigma'], ascending=False)

plot = sns.lineplot(x=sigma, y = "Acc", data=best_miu, hue="Matrix", markers=True)
plot.figure.set_figwidth(9)
plot.figure.set_figheight(7)

for item in plot.get_xticklabels():
    item.set_rotation(45)
    
plot.set_xlabel("Sigma", fontsize = 15)
plot.set_ylabel("Accuracy", fontsize = 15)

plt.savefig('sigma.pdf', dpi=300)
plt.show()
```

```python
best_sig = data_pd[data_pd['Sigma']==0.2]
sigma = [ f"{i}" for i in sorted(best_sig['Alpha'], reverse=True)]


best_sig = best_sig.sort_values(by=['Alpha'], ascending=False)

plot = sns.lineplot(x=sigma, y = "Acc", data=best_sig, hue="Matrix", markers=True)

plot.figure.set_figwidth(8.27)
plot.figure.set_figheight(5.5)

for item in plot.get_xticklabels():
    item.set_rotation(45)
    
plot.set_xlabel("Alpha", fontsize = 15)
plot.set_ylabel("Accuracy", fontsize = 15)

plt.savefig('alpha.pdf', dpi=300)
plt.show()
```
